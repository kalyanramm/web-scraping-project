```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
import re
from decimal import Decimal
import matplotlib.dates as mdates
from matplotlib.dates import date2num 

%matplotlib inline
```

```python
df = pd.read_csv('benchmarks.csv', header=None)
df.columns = ['cpu_name', 'price', 'cpu_mark', 'cpu_value', 'thread_mark', 'thread_value', 'tdp_w', 'power_perf', 'test_date', 'socket', 'category', 'clock_speed', 'turbo_speed', 'cores_threads', 'rank', 'samples']
df = df.replace('Unknown', np.nan)
df = df.replace('', np.nan)

df['test_date'] = pd.to_datetime(df['test_date'], errors='coerce')
df = df.dropna(subset=['test_date'])
df = df.sort_values('test_date')
df = df.set_index('test_date')

df['cpu_company'] = list(map(lambda x: 'Amd' if 'AMD' in x else 'Intel' if 'Intel' in x else np.NaN, df['cpu_name']))
df['clock_speed'] = list(map(lambda x: x.replace('Clock Speed: ', ''), df['clock_speed']))
df['turbo_speed'] = list(map(lambda x: x.replace('Turbo Speed: ', ''), df['turbo_speed']))
df['samples'] = list(map(lambda x: x.replace('Samples: ', ''), df['samples']))
df['rank'] = list(map(lambda x: Decimal(x.replace('Rank: ', '')), df['rank']))
df['price'] = pd.to_numeric(list(map(lambda x: Decimal(str(x).strip('$').strip('*').replace(',', '')), df['price'])))
df['cpu_mark'] = pd.to_numeric(list(map(lambda x: Decimal(str(x).replace(',', '')), df['cpu_mark'])))
df['thread_mark'] = pd.to_numeric(list(map(lambda x: Decimal(str(x).replace(',', '')), df['thread_mark'])))
df['power_perf'] = pd.to_numeric(list(map(lambda x: Decimal(str(x).replace(',', '')), df['power_perf'])))

df.to_csv('benchmarks_sanitized.csv')
df.head()
```

```python
df_price = df[['cpu_company', 'price']].dropna()

ax = sns.boxplot(x='cpu_company', y='price', data=df_price)
ax.set(xlabel='Company', ylabel='Price', ylim=(0, 5000))
plt.show()
```

```python
#Desktop

df_desk = df[df.category=='Desktop'].copy().reset_index()
df_desk = df_desk.dropna(subset=['cpu_mark', 'cpu_company'])
df_desk['test_date'] = df_desk['test_date'].apply(date2num) 

sns.lmplot('test_date', 'cpu_mark', df_desk, hue='cpu_company', palette="Set2", col='cpu_company')
```
